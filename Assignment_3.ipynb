1. Write a program that takes a list of numbers and prints all prime numbers in the list using loops. Do not use any built-in prime-checking functions.

#### ✅ What is a Prime Number?

A prime number is a number:

- Greater than 1

- Divisible only by 1 and itself

Examples: 2, 3, 5, 7, 11…


### ✅ Algorithm (Step-by-Step Logic)

1. Start

2. Take a list of numbers

3. For each number in the list:

    - If number ≤ 1 → skip (not prime)

    - Otherwise:

        - Assume it is prime

        - Check divisibility from 2 to number−1

        - If divisible by any → mark as NOT prime

4. If still marked prime → print it

5. Repeat for all numbers

6. Stop

```python
number = [1,2,3,4,5,10,13,17,20,23]

print("Prime numbers in the list:")

for num in number:

    if num> 1:
        is_prime = True

        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            print(num)
```

3. Write a user-defined function that accepts a list of integers and returns: 
• Count of even numbers 
• Sum of odd numbers 
Use loops and conditional statements inside the function. 

#### ✅ Algorithm

1. Start

2. Define a function

3. Initialize:

    - even_count = 0

    - odd_sum = 0

4. For each number in the list:

    - If number is even → increase even_count

    - Else → add number to odd_sum

5. Return even_count and odd_sum

6. Stop

```python
def count_even_sum_odd(numbers):
          
    even_count = 0
    odd_sum = 0

    for num in numbers:

        if num % 2 == 0:
            even_count += 1
        else:
            odd_sum += num

    return even_count, odd_sum


# Example usage
nums = [2, 5, 8, 3, 10, 7]

even, odd = count_even_sum_odd(nums)

print("Count of even numbers:", even)
print("Sum of odd numbers:", odd)
```

4. Create a function that calculates simple interest. Use default arguments for rate and time. Demonstrate calling the function using both positional and keyword arguments.


#### ✅ Simple Interest Formula 
$$
SI = \frac{P \times R \times T}{100}
$$
  

#### 🎯 Important

Use:

- ($$ ... $$ → For big centered formula (like your image))

- \frac{numerator}{denominator} → For fraction

- \times → For multiplication symbol


Where:

P = Principal

R = Rate (default)

T = Time (default)


### ✅ Algorithm

1. Start

2. Define function with parameters:

	- principal

	- rate (default = 5)

	- time (default = 2)

3. Calculate simple interest:

	- SI = (P × R × T) / 100

4. Return SI

5. Call function using positional arguments

6. Call function using keyword arguments

7. Stop


```python
def simple_interest(principal, rate=5, time=2):

    si = (principal * rate * time) / 100
    return si


# Positional arguments
result1 = simple_interest(1000)

# Keyword arguments
result2 = simple_interest(principal=2000, rate=6, time=3)

print("Simple Interest (positional):", result1)
print("Simple Interest (keyword):", result2)
```

5. Create a Student class with attributes: name, marks. 
Add a method to calculate grade based on marks. 
Create at least two objects and display their grades. 

### ✅ Algorithm

1. Start

2. Define Student class

3. Create constructor to initialize name and marks

4. Define calculate_grade() method

5. Inside method:

    - If marks ≥ 90 → Grade A

    - Else if marks ≥ 75 → Grade B

    - Else if marks ≥ 60 → Grade C

    - Else if marks ≥ 40 → Grade D

    - Else → Grade F

6. Create two student objects

7. Call method and display grades

8. Stop



```python
class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def calculate_grade(self):

        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 60:
            return "C"
        elif self.marks >= 40:
            return "D"
        else:
            return "F"


# Creating objects
student1 = Student("Bhavesh", 85)
student2 = Student("Priya", 35)

# Display grades
print(student1.name, "Grade:", student1.calculate_grade())
print(student2.name, "Grade:", student2.calculate_grade())
```

6. Create a BankAccount class with private attributes for balance. 
Provide methods to deposit and withdraw money. Ensure balance 
cannot be accessed directly. 

### ✅ Algorithm

1. Start

2. Create BankAccount class

3. Declare private variable __balance

4. Initialize balance using constructor

5. Create deposit(amount):

    - Add amount to balance

6. Create withdraw(amount):

    - If amount ≤ balance → subtract

    - Else → display insufficient funds

7. Create get_balance() method

8. Create object

9. Call deposit and withdraw

10. Display balance

11. Stop

```python
class BankAccount:

    def __init__(self, balance):
        self.__balance = balance     # Private attribute

    def deposit(self, amount):
        self.__balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance


# Create object
account = BankAccount(1000)

account.deposit(500)
account.withdraw(300)

print("Current Balance:", account.get_balance())
```

7. Write a program that takes two numbers from the user and performs 
division. 
Handle the following exceptions: 
• Division by zero 
• Invalid input type 

### ✅ Concepts Used

- try

- except

- ZeroDivisionError

- ValueError

- finally (optional)

This is called Exception Handling.

### ✅ Algorithm

1. Start

2. Ask user to enter first number

3. Ask user to enter second number

4. Convert inputs to numbers

5. Try to divide first number by second

6. If second number is zero → handle ZeroDivisionError

7. If input is not numeric → handle ValueError

8. Display result if no error

9. Stop

```python
try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    result = num1 / num2

    print("Result:", result)

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except ValueError:
    print("Error: Invalid input! Please enter numeric values.")

finally:
    print("Program finished.")
```

8. Create a custom exception that raises an error if a user enters age less 
than 0 or greater than 120.

### ✅ Algorithm

1. Start

2. Create custom exception class InvalidAgeError

3. Ask user to enter age

4. Convert input to integer

5. If age < 0 or age > 120:

    - Raise InvalidAgeError

6. Else:

    - Display valid age

7. Catch exception and print error message

8. Stop

```python
# Custom Exception Class
class InvalidAgeError(Exception):
    pass


try:
    age = int(input("Enter your age: "))

    if age < 0 or age > 120:
        raise InvalidAgeError("Age must be between 0 and 120.")

    print("Valid Age Entered:", age)

except InvalidAgeError as e:
    print("Custom Exception:", e)

except ValueError:
    print("Please enter a valid integer age.")
```

10. Write a program that reads numbers from a file and calculates: 
• Total 
• Average 
Handle the case if the file does not exist. 

#### ✅ Algorithm

1. Start

2. Try to open file in read mode

3. Initialize:

    - total = 0

    - count = 0

4. Read each line from file

5. Convert line to number

6. Add number to total

7. Increase count

8. After loop:

    - Calculate average = total / count

9. Display total and average

10. Close file

11. If file not found → print error message

12. Stop

```python
try:
    file = open("numbers.txt", "r")

    total = 0
    count = 0

    for line in file:
        num = float(line.strip())
        total += num
        count += 1

    if count > 0:
        average = total / count
        print("Total:", total)
        print("Average:", average)
    else:
        print("File is empty.")

    file.close()

except FileNotFoundError:
    print("Error: File does not exist.")
```
